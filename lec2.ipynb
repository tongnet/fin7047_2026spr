```python
# python program to get bitcoin prices (close, high, low, open) from yahoo finance

import yfinance as yf

# Define the ticker symbol for CME Bitcoin futures
ticker = "BTC-USD"

# Download historical data
df = yf.download(ticker, period="1y", interval="1d")

print(df.head())
print(df.tail())
```
