```python
# Cumulative density
prob = {1:1/6, 2:1/6, 3:1/6, 4:1/6, 5:1/6, 6:1/6}
cdf = 0
for i in prob.keys():
    if i<=3:
        cdf += prob[i]
print(cdf)
```

```python
# CDF of A Normal distribution

import scipy
from scipy.stats import norm
x1=-1
x2=1
#cdf_probability = scipy.stats.norm.cdf(x, loc=0, scale=1)
y1 = scipy.stats.norm.cdf(x1)
y2 = scipy.stats.norm.cdf(x2)
cdf = y2-y1
print('The probability of X between -1 and 1 is', round(cdf,3))
```

```python
# calculate expected value
payoffs = {20000:0.01, 10000:0.02, 5000:0.05, 1000:0.07, 0:0.85}
mu1 = 0
for payoff, p in payoffs.items():
    mu1 += p*payoff
print(mu1)
```

```python
# standard deviation of a distribution before purchasing insurance

import math
payoff1 = {20000:0.01, 10000:0.02, 5000:0.05, 1000:0.07, 0:0.85}
mu1 = 0
for payoff, p in payoff1.items():
    mu1 += p*payoff

var1 = 0
for payoff, p in payoffs.items():
    var1 += p*(payoff - mu1)**2
print("The variance of costs before insurance is", var1)
std1 = math.sqrt(var1)
print("The standard deviation of costs before insurance is", std1)
```

```python
# standard deviation of a distribution after purchasing insurance

import math
payoff2 = {1800:0.85, 800:0.15}
mu2 = 0
for payoff, p in payoff2.items():
    mu2 += p*payoff

var2 = 0
for payoff, p in payoff2.items():
    var2 += p*(payoff - mu2)**2
print("The variance of costs after insurance is", var2)
std2 = math.sqrt(var2)
print("The standard deviation of costs after insurance is", std2)
```
